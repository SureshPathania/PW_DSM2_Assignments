```python
''' ----------------------- Flask Assignment ------------------------
Que 1 - What is Flask Framework? What are -----------------------
Ans - Flask is a web development framework written in Python. It provides tools, libraries and technologies that 
allows to build a web application. It is easy to learn and use. It is a WSGI(Web Server Gateway Interface) 
framework. Flask is considered a micro web framework. It is not dependent upon external libraries to perform the
tasks of a framework. 

Advantages of flask framework -
a) Lightweight: Flask is a lightweight framework because it is independent of external libraries and it gives a 
quick start for web development having complex applications.
b) Compatible: Flask is compatible with the latest technologies such as machine learning, agile development, cloud
technologies, etc.
c) Independent: Flask allows full control to the developers for creating web applications. A developer can do the
experiment with the libraries and architecture of the framework.
d) Integrated Unit Testing: Flask offers an integrated unit testing feature that helps in faster debugging, robust
development, and independence to do experiments.
e) Flexible and Scalable: Flask supports WSGI templates that help in flexibility and scalability in the web 
development process. 
g) Secure Cookies: Secure cookie is an attribute of an HTTP request that enables the security of channels and 
ensures no unauthorized person has access to the text. Flask supports the feature of secure cookies. 
h) It is suitable for single-page application.
i) Flask provides an in-built debugger that offers virtual debugging.
'''
```

```python
'''Que 2 - Create a simple Flask application to dispaly---------------------
Ans - 
'''
from flask import Flask
app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
'''Que 3 - What is App routing in Flask? Why -----------------
Ans -
App Routing - App Routing means mapping the URLs to a specific function that will handle the logic for that URL. 
App routing is the technique used to map the specific URL with the associated function intended to perform some 
task. Modern web frameworks use more meaningful URLs to help users remember the URLs and make navigation simpler. 
To bind a function to an URL path we use the app.route decorator.

Why we use app routes -
App routes are used to help users remember the application URLs.
It is helpful to access the desired page directly without navigating from the home page.
'''
```

```python
'''Que 4 - Create a "/welcome" route to display --------------------------
Ans -
'''
from flask import Flask
app = Flask(__name__)

@app.route("/")
def hello():
    return "<b>Company Name : ABC Corporation<br>Location : India<br> Contact Detail : 999-999-9999<br></b>"

@app.route("/welcome")
def welcome():
    return "<h3>Welcome to ABC Corporation</h3>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
'''Que 5 - What function is used in flask for URL binding? Write a ------------------------
Ans - There are two ways through which we can bind the URLs to their associated function:-
1.) Using route() decorator
2.) Using add_url_rule() function

1) route() decorator - The route() decorator in Flask is used to bind URL to a function. 
Syntax - 
            @app.route('/route-name')
e.g. − 
            @app. route('/hello') 
            def hello_world(): 
                return 'hello world' 

Here, URL '/hello' rule is bound to the hello_world() function.

2) add_url_rule() function - URL mapping can be done by using the add_url_rule() function. This approach is mainly 
used to import the view function from another module. In fact, the app.route() calls this function internally.
Syntax -
        add_url_rule(<url rule>, <endpoint>, <view function>)
e.g. - 
'''
# example of add_url_rule
from flask import Flask
app = Flask(__name__)

def root():
return "Coding World."

def hello():
return "Hello, Welcome to Coding World."

app.add_url_rule('/','root',root); 
app.add_url_rule('/hello','hello',hello); 

if __name__ == "__main__":
app.run(debug=True)

'''
url_for() function - The url_for() method, is used to prepare a URL, for a function dynamically, such that, 
changing URLs, in the application, is avoided. It accepts, the name of the view function, as the first argument, 
and, any number of keywords, to be sent (to the view function), as the second argument. 
Syntax: 
        url_for(‘<function_name>’,<key> = <value>)
e.g. -
'''
#example of url_for()
from flask import url_for, Flask
appFlask = Flask(__name__)
@appFlask.route('/home')
def home():
   return 'We are in Home Page!'
with appFlask.test_request_context():
    print(url_for('login'))
```
