```python
'''Que2 - Give advantages and -------------------------
Ans -
Advantages of using API -
a) Increased Interoperability: API solutions enable different software systems to communicate with each other, 
regardless of their programming language or platform. This enhances the interoperability between systems, allowing 
for the exchange of data and functionality.
b) Improved Efficiency: API driven infrastructures enable developers to reuse code and functionality from existing 
software systems, saving time and reducing the amount of code that needs to be written from scratch.
c) Better User Experience: APIs can enable third-party developers to create applications that integrate with 
existing software systems, providing a better user experience and increased functionality.
d) Increased Flexibility: APIs provide developers with greater flexibility, enabling them to choose the best 
programming languages and frameworks for their needs.
e) New Revenue Streams: APIs can enable companies to create new revenue streams by providing third-party developers
with access to their software systems.

Disadvantages of using API - 
a) Security Concerns: API infrastructures can be vulnerable to security breaches, such as SQL injection attacks and
cross-site scripting (XSS) attacks, making it crucial to implement security measures to protect against these 
threats.
b) Complexity: APIs can be complex, requiring a significant amount of development effort to implement and maintain.
c) Compatibility Issues: APIs can have compatibility issues between different versions of the same software system
or between different systems, which can lead to errors and system failures.
d) Dependency on Third-Party Developers: APIs rely on third-party developers to create applications that integrate
with the software system. This can create dependencies that are outside of the control of the company that owns the
software system.
e) Governance and Management: APIs require proper governance and management to ensure they are being used correctly
and consistently across different applications and systems.
'''
```

```python
'''Que 4 - Explain REST and SOAP ---------------------
Ans -
REST - REST means Representational State Transfer. Representational state transfer is a software architectural 
style that was created to guide the design and development of the architecture for the World Wide Web. REST 
defines a set of constraints for how the architecture of an Internet-scale distributed hypermedia system, such 
as the Web, should behave. REST is solid and elastic architecture. Therefore, it’s so popular and widely used in
many modern APIs.
It is a software architectural style that defines the set of rules to be used for creating web services.
Constraints for REST Architecture -
a) Uniform Interface
b) Client-Server Architecture
c) Statelessness
d) Cacheability
e) Layered Syatem
f) Code on Demand

SOAP - SOAP means Simple Object Access Protocol. SOAP is a network platform used in a web service to exchange or 
communicate data between two different machines on a network. It uses the XML format of data to transfer messages
over the HTTP protocol. In Web services, SOAP allows the user request to interact with other programming languages.
In this way, it provides a way to communicate between applications running on different platforms (OS), with 
programming languages and technologies used in web service.

Characteristics of SOAP -
a) It is an open standard protocol used in the web service to communicate via internet.
b) It is used to broadcast a message over the network.
c) It is used to call remote procedures and exchange documents.
d) It can be used on any platform and can support multi-languages. So, it is a platform and language independent.
e) It uses the XML format to send messages over the HTTP protocol.
f) The structure of a SOAP message consists of an envelope, header, and body element.

Shortcomings of SOAP architecture - 
a) SOAP used only XML format data in web service, whereas JSON and other lightweight formats are not supported
by it.
b) It is slow because it uses XML format, whereas the payload for a simple string message is large.
c) There are no security features in the SOAP specification.
d) There is no state reference for the remote object in the SOAP client.
'''
```
