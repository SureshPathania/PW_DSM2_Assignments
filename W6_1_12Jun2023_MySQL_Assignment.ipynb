```python
'''Que 2 - What is DDL? Explain ------------------------
Ans - DDL - DDL means Data Definition Language. It is a language to create and modify the structure of database
objects in a database. It is a subset of SQL. It is a language used for describing data and its  relationships 
in the database.

CREATE - CREATE is a command used to create databases, tables, triggers and other database objects.
Syntax - 
CREATE DATABASE database-name;
e.g. -
CREATE TABLE student( rollno int, name varchar(20), class varchar(15));

DROP - DROP command is used to delete or remove the database objects from the SQL database.
Syntax - 
DROP DATABASE database-name;
e.g. -
DROP TABLE student;

ALTER - ALTER command is used to change or modify the existing structure of the database. It also changes the 
schema of the database.
Syntax -
ALTER TABLE table-name ADD|DROP|MODIFY column-name [datatype];
e.g -
ALTER TABLE student ADD age int;
ALTER TABLE student DROP age;
ALTER TABLE student MODIFY name varchar(30);

TRUNCATE - TRUNCATE command is used to delete all records from a table. It will also remove the space allocated for
storing the records.
Syntax -
TRUNCATE TABLE table-name;
e.g. -
TRUNCATE TABLE student;
'''
```

```python
'''Que3 - What is DML? Explain ----------
Ans -
DML - DML means Data Manipulation Language. It is a language used to change the existing data in the database. It
is also a subset of SQL. DML commands help us to easily access, store, modify, update and delete the existing 
records in the database. 

INSERT - INSERT command is used to insert the data or records in the database tables.
Syntax - 
INSERT INTO table-name [(col1, col2,...coln)] VALUES(val1, val2, .... valn);
e.g. -
INSERT INTO student VALUES(11, "Sumit", 10);

UPDATE - UPDATE command is used to update or modufy the existing data in the database tables.
Syntax -
UPDATE table-name SET [col-1 = val-1,........ col-n = val-n] WHERE condition;
e.g. -
UPDATE student SET class = 9 WHERE rollno > 10;

DELETE - DELETE command is used to delete record(s) from the database tables.
Syntax -
DELETE FROM TABLE table-name WHERE condition;
e.g. -
DELETE FROM student WHERE class=10;
'''
```

```python
'''Que 4 - What is DQL? Explain ------------
Ans -
DQL - DQL means Data Query Language. It is also a subset of SQL. It is basically used to retrieve data
from the database tables. DQL statements are used for performing queries on the data within schema 
objects.

SELECT - Select command is used to extract the specific records from the database tables.

Syntax -
SELECT (col-1, col-2,.......col-n)|* FROM table-name WHERE condition GROUP By col-name HAVING condition; 
e.g. -
SELECT * FROM student;
SELECT rollno, name FROM student;
SELECT * FROM student WHERE class=9;
'''
```

```python
'''Que 5 - Explain Primary key and -------------------
Ans -
Primary Key - A primary key is the column or columns that contain values that uniquely identify each row
in the table. Primary key column cannot store duplicate values. Primary key column cannot be left blank or
cannot have NULL value. There can be only one primary key in a table.

Foreign Key - The foreign key is a group of one or more columns in a database to uniquely identify another
database record in some other table to maintain the referential integrity. It is also known as the 
referencing key that establishes a relationship between two different tables in a database. A foreign key 
always matches the primary key column in another table. It is used to link two tables together.
'''
```
