```python
'''Q 4 - Why LookupError class is used-----------------------
Ans - This is the base class for those exceptions that are raised when a key or index used on a mapping or sequence is invalid or not found. The exceptions raised are :
KeyError - KeyError is raised when the mapping key is not found in the set of existing keys.
IndexError - IndexError is raised when a sequence is referenced which is out of range.
'''
import logging
logging.basicConfig(filename="test1.log", level=logging.DEBUG, format="%(asctime)s %(message)s")

try: 
    a = [1, 2, 3] 
    logging.info("{}".format(a[3]))

except IndexError: 
    logging.error("Index out of bound error.")
    
else: 
    logging.info("Success")

try:
    dict_array = {'a':1, 'b':2, 'c':3,'d':4}
    logging.info("{}".format(dict_array['x']))
    
except KeyError:
    logging.error("Key not found in dictionary")

else: 
    logging.info("Success")

logging.info("Execution completed.")
logging.shutdown()
```
