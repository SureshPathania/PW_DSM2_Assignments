```python
'''Q 4 - Explain with example ---------------
Ans - 
try - try block is used to catch the exceptions. Program code is placed inside the try block to test it for errors.
If any exception occurs during the execution of the code, it will throw the exception in the except block to handle
the exception. 

else - else block is used to place the code that will be exceuted only in case when no exception has been raised in
the try block during execution of the program. If any exception is raised during the execution in try block, then 
else block will not be executed.

finally - finally block is a block which will be executed in all situations. So this block is used to place that 
code which has to be executed in all situations whether an exception is raised or not.

raise - raise keyword raises an exception. It is generally used to raise a user customized exception. 
'''
import logging
logging.basicConfig(filename="test.log", level=logging.DEBUG, format="%(asctime)s %(message)s")

class vote_validation(Exception):
    def __init__(self, msg):
        self.msg = msg

def validate_vote(age, name):
    if age < 0 :
        raise vote_validation("Age cannot be negative.")
    elif age < 18:
        raise vote_validation("Sorry {}, you are a minor. You cannot cast vote.".format(name))
    else:
        logging.info("Congrats {}, you are eligible to cast your vote.".format(name))

try:
    name = input("Enter your name = ")
    age = int(input("Enter your age : "))
    validate_vote(age, name)

except vote_validation as e:
    logging.error(e)
    
else:
    logging.info("Vote casted successfully.")
    
finally:
    logging.info("Execution completed.")
    logging.shutdown()
```

```python
'''Q 5 - What are custom exception in python?-------------
Ans - 
Custom Exceptions - A custom exception in python is a user defined exception that is created by the programmer.

Need for Custom Exceptions - Custom exceptions are important tool for handling errors related to application 
specific requirements. Custom exceptions help in making code readable and maintainable, enforcing design patterns,
precise error reporting and enhancing code documentation.
'''
import logging
logging.basicConfig(filename="test.log", level=logging.DEBUG, format="%(asctime)s %(message)s")

class FahrenError(Exception):
    def __init__(self, msg):
        self.msg = msg

def validate_fahren(f):
    if f<32 or f>212:
        raise FahrenError("{} not in the range(32,212)".format(f))
    else:
        logging.info("{} is in valid range.".format(f))

try:
    fah = int(input("Enter temprature in fahrenheit : "))
    validate_fahren(fah)
    cel = (fah - 32) * 5 / 9
    logging.info("{} fahrenheit is equivalent to {} celcius.".format(fah,cel))
    
except FahrenError as e:
    logging.error(e)

finally:
    logging.info("Execution completed.")
    logging.shutdown()
```
